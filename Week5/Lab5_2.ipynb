**5.2: Course Enrollment & Collaboration (Set Focus)**

```python
# --- ข้อที่ 1 & 2: กำหนด Set ของนักศึกษาในแต่ละรายวิชา ---
# กำหนดเซตของนักศึกษาวิชา Python และ Java
python_students = {"S001", "S003", "S005", "S007", "S009"}
java_students = {"S002", "S003", "S006", "S007", "S010"}

print(f"Python Students: {python_students}")
print(f"Java Students:   {java_students}\n")

# --- ข้อที่ 3: หานักศึกษาที่เรียนทั้ง 2 วิชา (Intersection) ---
both_courses = python_students.intersection(java_students)
print(f"Students in Both Courses: {both_courses}")

# --- ข้อที่ 4: หานักศึกษาทั้งหมดแบบไม่ซ้ำกัน (Union) ---
all_unique_students = python_students.union(java_students)
print(f"All Unique Students: {all_unique_students}")

# --- ข้อที่ 5: หานักศึกษาที่เรียนเฉพาะวิชา Python เท่านั้น (Difference) ---
only_python = python_students.difference(java_students)
print(f"Students Only in Python: {only_python}")

# --- ข้อที่ 6: หานักศึกษาที่เรียนเพียงวิชาใดวิชาหนึ่งเท่านั้น (Symmetric Difference) ---
exclusive_students = python_students.symmetric_difference(java_students)
print(f"Students in Only One Course: {exclusive_students}\n")

# --- ข้อที่ 7: การลบข้อมูลซ้ำออกจาก List ---
all_students_list = ["S001", "S003", "S005", "S001", "S002"]
print(f"Original List (with duplicates): {all_students_list}")

# แปลง List เป็น Set เพื่อลบข้อมูลซ้ำ แล้วแปลงกลับเป็น List
unique_students_list = list(set(all_students_list))
print(f"Unique List (after removing duplicates): {unique_students_list}")
```

**LAB Challenge 5.2**

```python
# ============================================================
# LAB Challenge 5.2 - Student Club Management
# Set Focus
# ============================================================

ai_club = {"S001", "S002", "S003", "S004"}
robot_club = {"S003", "S004", "S005", "S006"}
game_club = {"S001", "S005", "S007", "S008"}


while True:

    print("\n===== Club Manager =====")
    print("1. Show AI Club Members")
    print("2. Show Robot Club Members")
    print("3. Show Game Club Members")
    print("4. Find Student")
    print("5. Students in All Clubs")
    print("6. Compare Clubs")
    print("7. Exit")

    choice = input("Choose menu: ")


    # แสดงสมาชิก AI
    if choice == "1":

        print("AI Club:", ai_club)


    # แสดงสมาชิก Robot
    elif choice == "2":

        print("Robot Club:", robot_club)


    # แสดงสมาชิก Game
    elif choice == "3":

        print("Game Club:", game_club)


    # ค้นหานักศึกษา
    elif choice == "4":

        student = input("Enter Student ID: ")

        print("\nMember of:")

        found = False

        if student in ai_club:
            print("- AI Club")
            found = True

        if student in robot_club:
            print("- Robot Club")
            found = True

        if student in game_club:
            print("- Game Club")
            found = True


        if not found:
            print("Not found.")


    # คนที่อยู่ทุกชมรม
    elif choice == "5":

        all_clubs = (
            ai_club
            .intersection(robot_club)
            .intersection(game_club)
        )

        print("Students in All Clubs:", all_clubs)


    # เปรียบเทียบชมรม
    elif choice == "6":

        print("\nAI & Robot:")
        print(ai_club.intersection(robot_club))


        print("\nAI Union Robot:")
        print(ai_club.union(robot_club))


        print("\nAI Only:")
        print(ai_club.difference(robot_club))


        print("\nOnly One Club:")
        print(
            ai_club.symmetric_difference(robot_club)
        )


    # ออกจากโปรแกรม
    elif choice == "7":

        print("Exit program.")
        break


    else:

        print("Invalid choice.")
```
