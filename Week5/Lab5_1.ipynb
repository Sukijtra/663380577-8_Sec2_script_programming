**5.1: Simple Contact Book Manager (Dictionary Focus)**

```python
# สร้าง Dictionary ว่างสำหรับเก็บข้อมูลผู้ติดต่อ
contacts = {}

while True:
    # แสดงเมนูตัวเลือก
    print("\n=== Contact Book Manager ===")
    print("1. Add/Update Contact")
    print("2. View Contact Details")
    print("3. List All Contacts")
    print("4. Delete Contact")
    print("5. Exit")

    choice = input("Enter choice (1-5): ")

    # เมนูที่ 1: เพิ่มหรืออัปเดตข้อมูลผู้ติดต่อ
    if choice == '1':
        name = input("Enter contact name: ")
        phone = input("Enter phone number: ")
        email = input("Enter email: ")

        # เก็บข้อมูลลงใน contacts โดยใช้ name เป็น key
        contacts[name] = {
            "phone": phone,
            "email": email
        }

        print(f"Contact '{name}' saved successfully!")

    # เมนูที่ 2: ดูรายละเอียดผู้ติดต่อ
    elif choice == '2':
        name = input("Enter contact name to view: ")

        # ใช้ .get() เพื่อดึงข้อมูล
        info = contacts.get(name)

        if info:
            print(f"\nName: {name}")
            print(f"Phone: {info['phone']}")
            print(f"Email: {info['email']}")
        else:
            print("Contact not found.")

    # เมนูที่ 3: แสดงรายการผู้ติดต่อทั้งหมด
    elif choice == '3':
        if not contacts:
            print("No contacts available.")
        else:
            print("\n--- All Contacts ---")
            for name, details in contacts.items():
                print(f"Name: {name} | Phone: {details['phone']} | Email: {details['email']}")

    # เมนูที่ 4: ลบข้อมูลผู้ติดต่อ
    elif choice == '4':
        name = input("Enter contact name to delete: ")

        # ใช้ .pop() ลบข้อมูล
        removed_contact = contacts.pop(name, None)

        if removed_contact:
            print(f"Contact '{name}' deleted successfully!")
        else:
            print("Contact not found.")

    # เมนูที่ 5: ออกจากโปรแกรม
    elif choice == '5':
        print("Exiting Contact Book Manager. Goodbye!")
        break

    else:
        print("Invalid choice! Please enter a number from 1 to 5.")
```

2. name

3. phone

4. email

6. items

```python
# ============================================================
# LAB Challenge 5.1 - Advanced Contact Book Manager
# Dictionary Focus
# ============================================================

contacts = {}

while True:
    print("\n===== Contact Book Manager =====")
    print("1. Add/Update Contact")
    print("2. View Contact")
    print("3. List All Contacts")
    print("4. Delete Contact")
    print("5. Search by Phone")
    print("6. Count Contacts")
    print("7. Edit Email")
    print("8. Clear All Contacts")
    print("9. Exit")

    choice = input("Enter choice (1-9): ")

    # เพิ่มข้อมูล
    if choice == "1":
        name = input("Enter name: ")
        phone = input("Enter phone: ")
        email = input("Enter email: ")

        contacts[name] = {
            "phone": phone,
            "email": email
        }

        print("Contact saved successfully!")

    # ดูข้อมูล
    elif choice == "2":
        name = input("Enter name to view: ")

        info = contacts.get(name)

        if info:
            print("\nName :", name)
            print("Phone:", info["phone"])
            print("Email:", info["email"])
        else:
            print("Contact not found.")

    # แสดงทั้งหมด
    elif choice == "3":

        if not contacts:
            print("No contacts available.")

        else:
            print("\n--- All Contacts ---")

            for name, info in contacts.items():
                print(
                    f"Name: {name} | "
                    f"Phone: {info['phone']} | "
                    f"Email: {info['email']}"
                )

    # ลบข้อมูล
    elif choice == "4":

        name = input("Enter name to delete: ")

        if contacts.pop(name, None):
            print("Deleted successfully!")

        else:
            print("Contact not found.")

    # ค้นหาจากเบอร์
    elif choice == "5":

        phone = input("Enter phone number: ")

        found = False

        for name, info in contacts.items():

            if info["phone"] == phone:
                print("\nFound Contact")
                print("Name :", name)
                print("Email:", info["email"])
                found = True

        if not found:
            print("Phone not found.")

    # นับจำนวนผู้ติดต่อ
    elif choice == "6":

        print("Total Contacts:", len(contacts))

    # แก้ไข Email
    elif choice == "7":

        name = input("Enter name: ")

        if name in contacts:

            email = input("New email: ")

            contacts[name]["email"] = email

            print("Email updated.")

        else:
            print("Contact not found.")

    # ลบทั้งหมด
    elif choice == "8":

        contacts.clear()

        print("All contacts cleared.")

    # ออกโปรแกรม
    elif choice == "9":

        print("Exit program.")
        break

    else:
        print("Invalid choice.")
```
