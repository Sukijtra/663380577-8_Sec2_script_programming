**Activity 4.1: List vs Tuple**
สิ่งที่อยากให้ทำ : สถานการณ์ต่อไปนี้ แต่ละเหตุการณ์ควรใช้ list หรือ tuple เพราะอะไร

1.ข้อมูลอุณหภูมิประจำวันที่มีการบันทึกไว้ตลอดหนึ่งสัปดาห์

**ควรใช้** List
สามารถเพิ่ม แก้ไข หรือลบข้อมูลได้

2.ค่าสี RGB ของเฉดสีเฉพาะเฉดหนึ่ง (เช่น สีแดงคือ (255, 0, 0))

**ควรใช้** Tuple
ค่าคงที่ ไม่ควรเปลี่ยน

3.รายการสินค้าในรถเข็นชอปปิงที่สามารถกดเพิ่มหรือลบออกได้

**ควรใช้** List
เพิ่มและลบสินค้าได้

4.รายชื่อวันในหนึ่งสัปดาห์เรียงตามลำดับ (จันทร์ - อาทิตย์)

**ควรใช้** Tuple
ลำดับวันเป็นข้อมูลคงที่

5.คะแนนเก็บของนักเรียนจากการส่งงานชุดต่างๆ ซึ่งอาจมีการอัปเดตหรือแก้ไขคะแนนในภายหลังได้

**ควรใช้** List
คะแนนอาจมีการแก้ไข

6.พิกัด GPS ของสถานที่สำคัญหรือจุดแลนด์มาร์กที่ตั้งอยู่กับที่

**ควรใช้** Tuple
พิกัดเป็นข้อมูลคงที่

**Activity 4.2: Interactive To-Do List**

```python
todo_list = []

while True:
    print("\n===== TO-DO LIST =====")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Complete Task")
    print("4. Exit")

    choice = input("Choose: ")

    if choice == "1":
        task = input("Enter task: ")
        todo_list.append(task)
        print("Task added.")

    elif choice == "2":
        if len(todo_list) == 0:
            print("No tasks.")
        else:
            print("\nTasks:")
            for i in range(len(todo_list)):
                print(i, "-", todo_list[i])

    elif choice == "3":
        if len(todo_list) == 0:
            print("No tasks to remove.")
        else:
            for i in range(len(todo_list)):
                print(i, "-", todo_list[i])

            index = int(input("Enter task index to complete: "))

            if 0 <= index < len(todo_list):
                removed = todo_list.pop(index)
                print(removed, "completed!")
            else:
                print("Invalid index.")

    elif choice == "4":
        print("Goodbye!")
        break

    else:
        print("Invalid choice.")
```
