**Lab 4.2: Geometric Calculations (Tuple)**

```python
import math

point1 = (3, 4)
point2 = (6, 8)

print("Point 1: x =", point1[0], ", y =", point1[1])
print("Point 2: x =", point2[0], ", y =", point2[1])

# ทดลองแก้ไขค่า (Tuple เปลี่ยนค่าไม่ได้)
# point1[0] = 5

distance = math.sqrt((point2[0]-point1[0])**2 + (point2[1]-point1[1])**2)

print("Distance =", distance)
```

อธิบาย Immutability
Tuple เป็นข้อมูลแบบ Immutable จึงไม่สามารถแก้ไขสมาชิกภายในได้
หากเขียน point1[0] = 5 จะเกิด TypeError เพราะ Tuple ไม่อนุญาตให้เปลี่ยนค่า

```python
import math

x1 = float(input("Enter x1: "))
y1 = float(input("Enter y1: "))

x2 = float(input("Enter x2: "))
y2 = float(input("Enter y2: "))

point1 = (x1, y1)
point2 = (x2, y2)

distance = math.sqrt((point2[0]-point1[0])**2 + (point2[1]-point1[1])**2)

print("Distance =", distance)
```
