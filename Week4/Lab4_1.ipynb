**Lab 4.1: Student Management System (List Operations)**

```python
# Student Management System

student_names = []

# เพิ่มรายชื่อนักเรียน 3-5 คน
num = int(input("How many students (3-5): "))

for i in range(num):
    name = input(f"Enter student {i+1}: ")
    student_names.append(name)

print("\nCurrent student list:")
print(student_names)

# ค้นหารายชื่อ
find_name = input("\nEnter a student name to search: ")

if find_name in student_names:
    print(find_name, "found at index", student_names.index(find_name))
else:
    print("Student not found.")

# ลบรายชื่อ
remove_name = input("\nEnter a student name to remove: ")

if remove_name in student_names:
    student_names.remove(remove_name)
    print("Student removed.")
else:
    print("Student not found.")

print("Updated student list:")
print(student_names)

# เรียงลำดับ
student_names.sort()

print("\nSorted student list:")
print(student_names)

# นับจำนวน
print("Total students:", len(student_names))
```

**Challenge: Student Management Menu**

```python
student_names = []

while True:
    print("\n===== Student Management =====")
    print("1. Add Student")
    print("2. Remove Student")
    print("3. Find Student")
    print("4. Sort Students")
    print("5. Show Students")
    print("6. Exit")

    choice = input("Choose: ")

    if choice == "1":
        name = input("Enter student name: ")
        student_names.append(name)
        print("Added successfully.")

    elif choice == "2":
        name = input("Enter student name to remove: ")
        if name in student_names:
            student_names.remove(name)
            print("Removed successfully.")
        else:
            print("Student not found.")

    elif choice == "3":
        name = input("Enter student name to find: ")
        if name in student_names:
            print("Found at index", student_names.index(name))
        else:
            print("Student not found.")

    elif choice == "4":
        student_names.sort()
        print("Students sorted.")

    elif choice == "5":
        print("Student List:")
        print(student_names)
        print("Total:", len(student_names))

    elif choice == "6":
        print("Goodbye!")
        break

    else:
        print("Invalid choice.")
```
