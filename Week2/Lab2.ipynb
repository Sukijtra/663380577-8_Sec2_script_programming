```python

num = int(input("Enter a number: "))

if num < 0:
    text = f"{num} is negative"
elif num > 0:
    text = f"{num} is positive"
else:
    text = f"{num} is zero"

print(text)

if num % 2 == 0:
    print(f"{num} is even")
else:
    print(f"{num} is odd")
```

1. ค่าตัวแปรเริ่มต้นที่กำหนดให้ temp = 25

is_raining = True

2. การพิจารณาเงื่อนไขทีละขั้นตอน (Step-by-Step) ขั้นตอนที่ 1: ตรวจสอบเงื่อนไขในช่อง if

โค้ด: if temp > 20 and not is_raining:

แทนค่า: if 25 > 20 and not True:

ยุบเงื่อนไข: if True and False: ซึ่งผลลัพธ์รวมจะได้เป็น False

ผลลัพธ์ขั้นตอนที่ 1: เงื่อนไขเป็นเท็จ โปรแกรมข้ามไป ไม่พิมพ์คำว่า "Perfect day for a picnic!"

ขั้นตอนที่ 2: ตรวจสอบเงื่อนไขในช่อง elif

โค้ด: elif temp <= 20 or is_raining:

แทนค่า: elif 25 <= 20 or True:

ยุบเงื่อนไข: elif False or True: ซึ่งตัวดำเนินการ or ขอแค่มีอันใดอันหนึ่งเป็นจริง ผลลัพธ์รวมจะได้เป็น True

ผลลัพธ์ขั้นตอนที่ 2: เงื่อนไขเป็นจริง! โปรแกรมจะเข้ามาทำงานในบล็อกนี้ทันที และหยุดตรวจเงื่อนไขอื่นที่เหลือ

2.2: เกมผจญภัยตามหาขุมทรัพย์ในป่า

1. เนื้อเรื่อง (Story) คุณหลุดเข้ามาในป่าลึกลับและเจอทางแยก 3 ทาง: ทางซ้ายมีเสียงน้ำตก, ทางตรงเป็นถ้ำมืด, และทางขวามีแสงไฟสลัวๆ คุณต้องเลือกเส้นทางเพื่อตามหาขุมทรัพย์และเอาชีวิตรอด

2. ลำดับการทำงาน (Pseudocode) เริ่มต้นเกม -> แสดงเนื้อเรื่องและตัวเลือก

รับค่าจากผู้ใช้: choice = input("คุณจะไปทางไหน? (ซ้าย / ตรง / ขวา): ")

เงื่อนไขและการตัดสินใจ (IF-ELIF-ELSE):

IF choice == "ซ้าย"

ผลลัพธ์: คุณเดินไปตามเสียงน้ำตกแล้วเจอลำธารใสสะอาด ได้ดื่มน้ำจนหายเหนื่อยและพบหีบสมบัติทองคำโบราณตกอยู่ข้าง ๆ

(จบเกม: ชนะ)

ELIF choice == "ตรง"

ผลลัพธ์: คุณเดินเข้าไปในถ้ำมืด ทันใดนั้นมีฝูงค้างคาวบินสวนออกมา ทำให้คุณตกใจก้าวพลาดตกหลุมลึกลึกลับลึก

(จบเกม: แพ้)

ELIF choice == "ขวา"

ผลลัพธ์: คุณเดินตามแสงไฟไปจนเจอแคมป์ไฟของคนป่า พวกเขาต้อนรับคุณอย่างดีและแบ่งอาหารให้ทาน แต่ไม่มีสมบัติอะไรเลย

(จบเกม: รอดชีวิตแบบธรรมดา)

ELSE (ถ้าพิมพ์คำอื่นที่ไม่ใช่ ซ้าย/ตรง/ขวา)

ผลลัพธ์: คุณยืนลังเลไม่ยอมเลือก จนกระทั่งสัตว์ร้ายในป่าตามมาเจอคุณ

(จบเกม: แพ้)
