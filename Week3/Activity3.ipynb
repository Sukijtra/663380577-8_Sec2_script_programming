Buggy Snippet A : Infinite Loop

```python
# Buggy Snippet A

counter = 0

while counter < 5:
    print(counter)
    # Missing counter increment!
```

ข้อผิดพลาด Infinite Loop
สาเหตุ ไม่มีการอัปเดตตัวแปรควบคุมลูป
วิธีแก้ไข เพิ่มคำสั่งอัปเดต เช่น counter += 1

Buggy Snippet C : Misplaced Break

```python
for char in "hello":
    print(char)

    if char == 'l':
        print("Found 'l', stopping!")
        break
```

ใช้ break ผิดตำแหน่ง
ลูปหยุดก่อนแสดงผลตามต้องการ
ย้ายตำแหน่ง break หรือ print() ให้เหมาะสม

```python
print("--- Printing a 5x5 square of asterisks ---")

size = 5

for row in range(size):

    for col in range(size):
        print("*", end="")

    print()
```

```python
print("--- Printing a right-angled triangle ---")

height = 5

for row in range(1, height + 1):

    for col in range(row):
        print("*", end="")

    print()
```

```python
print("--- Printing an inverted right-angled triangle ---")

height = 5

for row in range(height, 0, -1):

    for col in range(row):
        print("*", end="")

    print()
```

```python
height = 5

for row in range(1, height + 1):

    for space in range(height - row):
        print(" ", end="")

    for star in range(2 * row - 1):
        print("*", end="")

    print()
```
